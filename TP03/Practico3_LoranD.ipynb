## Funcion para aplicar el método de Euler

```python

def mEuler (fx, x0, t0, tf, n):
    deltat=(tf-t0)/n
    x = [x0]
    t = [t0]
    for i in range(n):
        x.append(x[i]+fx(x[i])*deltat)
        t.append(t[i]+deltat)
    return x, t   
        

def fn(x): 
    import numpy as npy
    return npy.sin(x) # Ingresar la función
        
x, t=mEuler(fn, 0.2, 0, 10, 1000)


#print(x)

#Graficar
#y= fn(x)

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 4))
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Gráfico de la Función')
ax.grid()
ax.plot(t, x,'r', label= "f(t)")
ax.legend()
plt.show()
```

Problema 3: Usando el programa del ejercicio anterior, resolv ́e el problema de valor inicial
 ̇x = (t − x)/2

en el intervalo [0, 3] con x(0) = 1 y para los siguientes cuatro valores de h = 1, 0,5, 0,25, 0,125. Resolv ́e
exactamente el problema determinando su dominio de definici ́on y compar ́a el error final (en t = 3) en
los cuatro casos.
Donde: x(t) = t − 2 + c exp(−t/2) con c = 3.

```python
def mEulerD (fx, x0, t0, tf, delta):
    #deltat=(tf-t0)/n
    n=int((tf-t0)/delta)
    x = [x0]
    t = [t0]
    for i in range(n):
        x.append(x[i]+fx(x[i], t[i])*delta)
        t.append(t[i]+delta)
    return x, t   
        

def fn2(x, t): 
    import numpy as npy
    return (t - x)/2 # Ingresar la función
        
x1, t1=mEulerD(fn2, 1, 0, 3, 1.0)
x05, t05=mEulerD(fn2, 1, 0, 3, 0.5)
x25, t25=mEulerD(fn2, 1, 0, 3, 0.25)
x125, t125=mEulerD(fn2, 1, 0, 3, 0.125)


import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 5))
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Gráfico de la aproximación a la función')
ax.set_xlim(0, 3)
ax.set_ylim(0, 2)
ax.grid()
ax.plot(t1, x1,'r', label= "f(t) c/delta t 1")
ax.plot(t05, x05,'g', label= "f(t) c/delta t 0.5")
ax.plot(t25, x25,'b', label= "f(t) c/delta t 0.25")
ax.plot(t125, x125,'k', label= "f(t) c/delta t 0.125")
ax.legend()
plt.show()
```

```python
# compara el error final (en t = 3) en los cuatro casos. Donde: x(t) = t − 2 + c exp(−t/2) con c = 3.
import numpy as npy
xt3 = 3 - 2 + 3*npy.exp(-3/2)


x1_3=x1[t1.index(3)]
x05_3=x05[t05.index(3)]
x25_3=x25[t25.index(3)]
x125_3=x125[t125.index(3)]

print("Error relativo con delta-t de 1: ", round((abs(xt3-x1_3))/xt3*100,4), "%")
print("Error relativo con delta-t de 0.5: ", round((abs(xt3-x05_3))/xt3*100,4), "%")
print("Error relativo con delta-t de 0.25: ", round((abs(xt3-x25_3))/xt3*100,4), "%")
print("Error relativo con delta-t de 0.125: ", round((abs(xt3-x125_3))/xt3*100,4), "%")
```

Problema 4: La llamada ecuación logística
dN
dt = rN (1 − N/K)

describe el crecimiento autolimitado de una poblaci ́on dada (suponiendo que no interact ́ua con otras
especies y que tiene fuentes limitadas de alimentos). Fue propuesta por Verhulst en 1838 y permite
describir al menos cualitativamente varios fen ́omenos poblacionales observados en la naturaleza. En esta
ecuaci ́on N (t) es el n ́umero de individuos de la colonia al tiempo t y K es una constante positiva.
Una soluci ́on N ∗ se dice estacionaria si satisface dN ∗/dt = 0, y por ende no cambia en el tiempo.
Observando la ecuaci ́on (2) es f ́acil verficar que solo existen dos soluciones estacionarias: N1 = 0 y
N2 = K.
Determin ́a el dominio de definici ́on de la ecuaci ́on y cu ́al de las dos soluciones estacionarias es estable y
cu ́al es inestable resolviendo num ́ericamente la ecuaci ́on diferencial con el m ́etodo de Euler para r = 2,
K = 100, en el intervalo 0 ≤ t ≤ 50 con h = 0,1 y considerando cinco condiciones iniciales diferentes:
a) N (0) = 0
b) N (0) = 2
c) N (0) = 50
d) N (0) = 120
e) N (0) = 200.

```python
def fn3(N): 
    return 2*N*(1 - N/100) # Ingresar la función


def mEuler (fx, x0, t0, tf, n):
    deltat=(tf-t0)/n
    x = [x0]
    t = [t0]
    for i in range(n):
        x.append(x[i]+fx(x[i])*deltat)
        t.append(t[i]+deltat)
    return x, t 

n=int(50/0.1)


N0, t = mEuler(fn3, 0, 0, 50, n)
N2, t = mEuler(fn3, 2, 0, 50, n)
N50, t = mEuler(fn3, 50, 0, 50, n)
N100, t = mEuler(fn3, 100, 0, 50, n)
N120, t = mEuler(fn3, 120, 0, 50, n)
N200, t = mEuler(fn3, 200, 0, 50, n)


import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 5))
ax.set_xlabel('t')
ax.set_ylabel('N')
ax.set_title('Gráfico de la aproximación a la función según N inicial (N0)')
ax.set_xlim(0, 6)
#ax.set_ylim(0, 2)
ax.grid()
ax.plot(t, N0,'r--', label= "f(t) con N0 =0")
ax.plot(t, N2,'k', label= "f(t) con N0 =2")
ax.plot(t, N50,'g', label= "f(t) con N0 =50")
ax.plot(t, N100,'b--', label= "f(t) con N0 =100")
ax.plot(t, N120,'b', label= "f(t) con N0 =120")
ax.plot(t, N200,'r', label= "f(t) con N0 =200")
ax.legend()
plt.show()
```
